# Measurements

```python
measurements = df_info.loc[0, "Measurements"]
print(type(measurements)) 
measurements
```

```python
measurements = measurements.replace("'",'"')
measurements
```

```python
json_measure = json.loads(measurements)
print(type(json_measure))
json_measure
```

```python
df_info["Measurements"] = df_info["Measurements"].str.replace("'",'"') 
df_info["Measurements"] = df_info["Measurements"].apply(json.loads)
df_info["Measurements"].head()
```

```python
h_w = df_info["Measurements"].apply(pd.Series)
```

```python
h_w
```

```python
df_info = pd.concat((df_info, h_w), axis=1)
```

```python
df_info.head()
```

```python
df_info = df_info.drop(columns=['Measurements'])
```

```python
df_info.info()
```

```python
df_info['Weight'] = df_info['Weight'].str.replace('kg', '')
```

```python
df_info['Weight'] = pd.to_numeric(df_info['Weight'])
```

```python
df_info.head()
```

```python
df_info['Height'] = df_info['Height'].str.replace('cm', '')
df_info['Height']= pd.to_numeric(df_info['Height'])
df_info.head()
```

```python
df_info = df_info.rename(columns={'Height': 'Height_cm', 'Weight': 'Weight_kg'})
```

```python
powers = df_powers.loc[0, "Powers"].split(",")
print(type(powers)) 
powers
```

```python
df_powers['power_split'] = df_powers['Powers'].str.split(",")
```

```python
df_powers.head()
```

```python
explode = df_powers.explode('power_split')
```

```python
df_powers = df_powers.drop(columns=['Powers', 'power_split'])
```
